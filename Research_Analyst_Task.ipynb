## Bar Graph for AI Adoption Rates

```python
import matplotlib.pyplot as plt

# Bar Graph: AI Adoption Rates in Different Sectors
sectors = ['Finance', 'Healthcare', 'Education']
adoption_rates = [40, 35, 25]

plt.figure(figsize=(6, 4))
plt.bar(sectors, adoption_rates, color=['blue', 'green', 'red'])
plt.title('AI Adoption Rates in Different Sectors (2024)')
plt.xlabel('Sectors')
plt.ylabel('Adoption Rate (%)')
```

## Pie Chart for Distribution of AI Applications in Finance

```python
# Pie Chart: Distribution of AI Applications in Finance (2024)
labels = ['Digital Banking', 'Fraud Detection', 'Investment Analysis', 'Customer Support', 'Others']
sizes = [30, 25, 20, 15, 10]

plt.figure(figsize=(6, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of AI Applications in Finance (2024)')

```

## Stacked Bar Graph for AI Implementation in Major Indian Cities


```python
import numpy as np

# Stacked Bar Graph: AI Implementation in Major Indian Cities Across Sectors
cities = ['Mumbai', 'Bangalore', 'Hyderabad', 'Delhi']
finance = [50, 40, 30, 20]
healthcare = [30, 35, 25, 20]
education = [20, 25, 15, 10]

ind = np.arange(len(cities))
width = 0.3

plt.figure(figsize=(6, 4))
plt.bar(ind - width, finance, width, label='Finance', color='blue')
plt.bar(ind, healthcare, width, label='Healthcare', color='green')
plt.bar(ind + width, education, width, label='Education', color='red')

plt.xlabel('Cities')
plt.ylabel('Number of AI Startups/Projects')
plt.title('AI Implementation in Major Indian Cities Across Sectors (2024)')
plt.xticks(ind, cities)
plt.legend()
```

## Line Graph for Projected AI Investment Growth

```python
# Line Graph: Projected AI Investment Growth (2024-2025)
years = ['2024', '2025']
finance_investment = [2.5, 3.2]
healthcare_investment = [1.8, 2.5]
education_investment = [1.2, 1.6]

plt.figure(figsize=(6, 4))
plt.plot(years, finance_investment, marker='o', label='Finance', color='blue')
plt.plot(years, healthcare_investment, marker='o', label='Healthcare', color='green')
plt.plot(years, education_investment, marker='o', label='Education', color='red')

plt.title('Projected AI Investment Growth (2024-2025)')
plt.xlabel('Year')
plt.ylabel('Investments (in Billion USD)')
plt.legend()

```
